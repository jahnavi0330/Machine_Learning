```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**data**

```python
df = pd.read_csv('breast-cancer.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

**we have categorical data in diagonasis column,so we have to convert them into numerical data**

```python
df['diagnosis'] = df['diagnosis'].map({'M':0,'B':1}).astype(int)
```

```python
df.sample(10)
```

**as id and unnamed columns are useless,we can remove them**

```python
df=df.drop(['id','Unnamed: 32'],axis=1)
df.head(3)
```

**splitting data**

```python
X = df.iloc[: ,1:]
y = df.iloc[: ,0]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=54)
len(x_test)
```

**introducing algorithm**

```python
from sklearn.neighbors import KNeighborsClassifier
cls=KNeighborsClassifier()
```

```python
cls.fit(x_train,y_train)
```

**testing model with trained data**

```python
y_train_pred=cls.predict(x_train)
len(y_train_pred)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
confusion_matrix(y_train,y_train_pred)
```

```python
accuracy_score(y_train,y_train_pred)
```

```python
cls.score(x_train,y_train)
```

**testing model with test data**

```python
y_test_pred = cls.predict(x_test)
```

```python
confusion_matrix(y_test,y_test_pred)
```

```python
accuracy_score(y_test,y_test_pred)
```

```python
cls.score(x_test,y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_test_pred))
```
