**Gradient Descent**

If the training data is good,after training we get prediction points close to the actual points los will be low and accuracy is good.
But,in case of a bad training data we get prediction points far to the actual points.At that time,we need to use the gradient descent optimizer.

gradient descent find MSE by applying derivation.
-> step_size = C.F + Learning rate
where,    C.F = cost function=one of the loss functions
          Learning rate = decides the speed to converge he point to global
                                                                   minima point

1.new slope = old_slope+step_size
2.new_intercept = old_intercept+step_size

**sample implementaton**

```python
Xs_train = np.array([1,2,3,4,5])
ys_train = np.array([5,6,8,3,9])

n = len(Xs_train)
m_value=0
c_value=0
iters=1000
learn_rate=0.04
for i in range(iters):
  ys_train_pred = m_value*Xs_train + c_value
  cost_fun = (1/n)*sum([pow(j,2) for j in ys_train - ys_train_pred])
  step_size = cost_fun+learn_rate
  m_der = -(2/n) * sum(Xs_train,(ys_train-ys_train_pred))
  c_der = -(2/n) * sum(ys_train,ys_train_pred)
  m_value = m_value - learn_rate*m_der
  c_value = c_value - learn_rate*c_der
  print(f'm : {m_value} and c : {c_value} and iterations : {i}')
```
