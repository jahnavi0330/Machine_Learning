```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Mall_Customers.csv')
```

```python
df = df.iloc[:,3:]
df.head(5)
```

```python
df.shape
```

```python
from sklearn.cluster import KMeans
```

```python
clu = KMeans(n_clusters=3)
```

```python
X=df.iloc[:,:]
```

```python
clu.fit(X)
```

```python
centroids = clu.cluster_centers_
```

```python
centroids
```

```python
clu.labels_
```

```python
 clu.labels_ == 1,0
```

```python
centroids[: ,1]
```

```python
plt.figure(figsize=(5,3))
plt.scatter(X.iloc[clu.labels_ == 0, 0], X.iloc[clu.labels_ == 0, 1],c='red')
plt.scatter(X.iloc[clu.labels_ == 1, 0], X.iloc[clu.labels_ == 1, 1],c='blue')
plt.scatter(X.iloc[clu.labels_ == 2, 0], X.iloc[clu.labels_ == 2, 1],c='green')
plt.scatter(centroids[: , 0] , centroids[: , 1],marker='*',color='black')
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
if clu.predict([[12,120]])[0] == 0:
  print(f'Red Group')
elif clu.predict([[420,180]])[0] == 1:
  print(f'Blue Group')
else:
  print(f'Green Group')
```

**Best_K**

```python
from sklearn.cluster import KMeans
```

```python
k_values = np.arange(2,15)
k_values
```

```python
wcss_vals=[]
for i in k_values:
  clu = KMeans(n_clusters=i)
  clu.fit(X)
  wcss_vals.append(clu.inertia_)
wcss_vals
```

```python
plt.figure(figsize=(5,3))
plt.title('Best_k_value')
plt.xlabel('k_values')
plt.ylabel('wcss value for each k_value')
plt.plot(k_values,wcss_vals,color='r',marker='*')
plt.show()
```
